```python
import pandas as pd

file_path = '/content/drive/MyDrive/StressLevelDataset.csv'
df = pd.read_csv(file_path)

print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of stress levels
plt.figure(figsize=(8, 5))
sns.histplot(df['stress_level'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Stress Levels', fontsize=14, weight='bold')
plt.xlabel('Stress Level', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Calculate correlation matrix for stress indicators
correlation_matrix = df.corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Between Stress Indicators', fontsize=16, weight='bold')
plt.show()
```

```python
# Create a 1x2 grid for bar plots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Bar plot for sleep quality vs stress level
sns.barplot(x='sleep_quality', y='stress_level', data=df, ax=axes[0], ci='sd', palette='Blues')
axes[0].set_title('Stress Level vs Sleep Quality', fontsize=14, weight='bold')
axes[0].set_xlabel('Sleep Quality (Score)', fontsize=12)
axes[0].set_ylabel('Average Stress Level', fontsize=12)

# Bar plot for blood pressure vs stress level
sns.barplot(x='blood_pressure', y='stress_level', data=df, ax=axes[1], ci='sd', palette='Oranges')
axes[1].set_title('Stress Level vs Blood Pressure', fontsize=14, weight='bold')
axes[1].set_xlabel('Blood Pressure (Score)', fontsize=12)
axes[1].set_ylabel('Average Stress Level', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Create a 2x2 grid for mental health, social support, peer pressure, and bullying
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Violin plot for mental health history vs stress level
sns.violinplot(x=df['depression'], y=df['stress_level'], ax=axes[0, 0], palette='Blues')
axes[0, 0].set_title('Stress Level vs Depression', fontsize=14, weight='bold')
axes[0, 0].set_xlabel('Depression (Score)', fontsize=12)
axes[0, 0].set_ylabel('Stress Level', fontsize=12)

# Violin plot for social support vs stress level
sns.violinplot(x=df['teacher_student_relationship'], y=df['stress_level'], ax=axes[0, 1], palette='Greens')
axes[0, 1].set_title('Stress Level vs Teacher-Student Relationship', fontsize=14, weight='bold')
axes[0, 1].set_xlabel('Teacher-Student Relationship (Score)', fontsize=12)
axes[0, 1].set_ylabel('Stress Level', fontsize=12)

# Violin plot for peer pressure vs stress level
sns.violinplot(x=df['future_career_concerns'], y=df['stress_level'], ax=axes[1, 0], palette='Oranges')
axes[1, 0].set_title('Stress Level vs Future Career Concerns', fontsize=14, weight='bold')
axes[1, 0].set_xlabel('Future Career Concerns (Score)', fontsize=12)
axes[1, 0].set_ylabel('Stress Level', fontsize=12)

# Violin plot for bullying vs stress level
sns.violinplot(x=df['noise_level'], y=df['stress_level'], ax=axes[1, 1], palette='Reds')
axes[1, 1].set_title('Stress Level vs Noise Level', fontsize=14, weight='bold')
axes[1, 1].set_xlabel('Noise Level (Score)', fontsize=12)
axes[1, 1].set_ylabel('Stress Level', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Identify the top 10% of individuals based on stress level
top_10_stress_threshold = df['stress_level'].quantile(0.90)
high_stress_individuals = df[df['stress_level'] >= top_10_stress_threshold]
rest_of_population = df[df['stress_level'] < top_10_stress_threshold]

# Compare key characteristics of high-stress vs rest of population
comparison_high_stress = pd.DataFrame({
    'Avg Stress Level (Top 10%)': high_stress_individuals['stress_level'].mean(),
    'Avg Stress Level (Rest of Population)': rest_of_population['stress_level'].mean(),
    'Avg Anxiety Level (Top 10%)': high_stress_individuals['anxiety_level'].mean(),
    'Avg Anxiety Level (Rest of Population)': rest_of_population['anxiety_level'].mean(),
    'Avg Depression Level (Top 10%)': high_stress_individuals['depression'].mean(),
    'Avg Depression Level (Rest of Population)': rest_of_population['depression'].mean(),
    'Avg Sleep Quality (Top 10%)': high_stress_individuals['sleep_quality'].mean(),
    'Avg Sleep Quality (Rest of Population)': rest_of_population['sleep_quality'].mean(),
    'Avg Blood Pressure (Top 10%)': high_stress_individuals['blood_pressure'].mean(),
    'Avg Blood Pressure (Rest of Population)': rest_of_population['blood_pressure'].mean(),
}, index=['Value']).T

# Create a bar plot to compare stress indicators for high-stress vs rest
plt.figure(figsize=(12, 6))
sns.barplot(x=comparison_high_stress.index, y=comparison_high_stress['Value'],palette='Set2')
plt.title('High-Stress vs Rest of Population: Stress Indicators', fontsize=16, weight='bold')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Average Value', fontsize=12)
plt.show()
```
